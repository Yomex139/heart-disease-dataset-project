# importing library

```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline
```

# importing dataset

```python
df = pd.read_csv('heart-disease.csv')
```

```python
df.head()
```

# ploting frame

```python
#changing ploting style
style.use('seaborn')
#preparing graph
fig,(ax0,ax1) = plt.subplots(nrows=2,
                             ncols=1,
                             sharex=True,
                            figsize=(12,12))
#imputing data into row1
scatter = ax0.scatter(df.age,
                     df.chol,
                     c=df.target,
                     cmap = 'winter')
#customizing row1
ax0.set(
       ylabel='CHOLESTEROL',
       title='Cholesterol Level')
#adding legeng to row1
ax0.legend(*scatter.legend_elements(),
          title='target')
#adding mean line(horizontal)
ax0.axhline(df.chol.mean(),
           linestyle = 'dashdot',
           c='orange')

#imputing data into row2
scatter = ax1.scatter(df.age,
                     df.thalach,
                     c=df.target,
                     cmap='winter')
#adding legend to row2
ax1.legend(*scatter.legend_elements(),
          title='target'),
#adding horizontal line
ax1.axhline(df.thalach.median(),
           linestyle = 'dashdot',
           c='red')
#customizing plot
ax1.set(xlabel='AGE',
       ylabel='THALACH',
       title='Max Heart Rate')
#adding general title
plt.suptitle(t='MEDICAL',fontsize=20,fontweight='bold')
ax1.set_ylim([60,220])
```
